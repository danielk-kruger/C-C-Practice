## Age Groups

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import random
```

```python
# Generate random ages between 20 and 80
ages = [random.randint(20, 80) for _ in range(100)]

# Generate random genders (F or M)
genders = [random.choice(['F', 'M']) for _ in range(100)]

data = {'Ages': ages, 'Sex': genders}

df = pd.DataFrame(data)
```

```python
# Create separate dataframes for each gender
female_df = df[df['Sex'] == 'F']
male_df = df[df['Sex'] == 'M']
```

```python
print(df.head())
```

```python
# Create histograms for each gender
plt.figure(figsize=(10, 6))
```

```python
plt.subplot(1, 2, 1)
plt.hist(female_df['Ages'], bins=10, edgecolor='black', alpha=0.7, color='pink')
plt.title('Age Distribution - Females')
plt.xlabel('Age')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2)
plt.hist(male_df['Ages'], bins=10, edgecolor='black', alpha=0.7, color='lightblue')
plt.title('Age Distribution - Males')
plt.xlabel('Age')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
print(round(df.describe().mean()['Ages']))
```

```python
# Set the style of seaborn
sns.set(style="whitegrid")
```

```python
# Create a density plot for each gender
plt.figure(figsize=(10, 6))
sns.kdeplot(data=df[df['Sex'] == 'F'], x="Ages", label="Female", fill=True)
sns.kdeplot(data=df[df['Sex'] == 'M'], x="Ages", label="Male", fill=True)

plt.xlabel("Age")
plt.ylabel("Density")
plt.title("Density Plot of Age Distribution by Gender")
plt.legend()
plt.show()
```
